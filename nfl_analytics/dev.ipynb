```python
from nfl_analytics.data import load_pandas
import pandas as pd
```

```python
df = load_pandas()
df.head()
```

```python
df[['game_id', 'old_game_id', 'year', 'week', 'posteam']].head(10)
```

```python
df.groupby(['year', 'posteam'])[['epa']].mean()
```

```python
# Sack yards are necessary to get accurate TEAM passing stats.
# Team passing yards are sum(passing_yards) - sum(sack_yards)
# Player passing stats are simply sum(passing_yards).
# Ultimately I want to account for sack. 
# Should either do team passing yards which bakes sacks in, or both sum of individual passing yards and sack yards.

# Create a new column 'sack_yards' and set it to pd.NA by default
df['sack_yards'] = pd.NA

# Set sack_yards to yards_gained for rows where sack is not equal to 0
df.loc[df['sack'] != 0, 'sack_yards'] = df['yards_gained']
df.head()
```

```python
# passing_yards matches individual player stats (all players' passing yards)
# does not match team passing stats (that subtracts sack yards)
df_agg = df.groupby(['game_id', 'posteam'])[['passing_yards', 'rushing_yards', 'yards_gained', 'sack_yards']].sum().reset_index()
df_agg

# Alternatively, could probably get team passing yards from yards_gained - rushing_yards.
```

```python
# Shows that the home/away score are the final score. and that every play 
# for a given game is the final score (as opposed to start/end of play score for example)
unique_home_scores = df[df['game_id'] == '2023_20_TB_DET']['home_score'].nunique()
print(f"Number of unique home scores for game_id 2023_20_TB_DET: {unique_home_scores}")
```

```python
# Group by game and team and combine offensive and defensive stats into single record

# Separate offensive and defensive stats
offensive_stats = df.groupby(['game_id', 'posteam'])[['passing_yards', 'rushing_yards', 'yards_gained', 'sack_yards']].sum().reset_index()
defensive_stats = df.groupby(['game_id', 'defteam'])[['passing_yards', 'rushing_yards', 'yards_gained', 'sack_yards']].sum().reset_index()

# Rename columns for defensive stats to distinguish them
defensive_stats.rename(columns={'defteam': 'team',
                                 'passing_yards': 'passing_yards_defense',
                                 'rushing_yards': 'rushing_yards_defense',
                                 'yards_gained': 'yards_gained_defense',
                                 'sack_yards': 'sack_yards_defense'}, inplace=True)

# Merge offensive and defensive stats
df_game = pd.merge(offensive_stats, defensive_stats, left_on=['game_id', 'posteam'], right_on=['game_id', 'team'])

# Drop redundant column
df_game.drop(['posteam'], axis=1, inplace=True)

df_game.tail()

```

```python
# sets score differential to last value for each game and team
# total points would probably be a good stat to get points per 
# game but this would probably require identifying team by home/away
df_game['score_differential_post'] = df.groupby(['game_id', 'posteam'])[['score_differential_post']].last().reset_index(drop=True)
# df_game[['score_differential_post', 'spread_line']] = df.groupby(['game_id', 'posteam'])[['score_differential_post', 'spread_line']].last().reset_index(drop=True)

# Calculate mean for the 'epa' column
df_game['mean_epa'] = df.groupby(['game_id', 'posteam'])['epa'].mean().reset_index(drop=True)

df_game.tail()
```
